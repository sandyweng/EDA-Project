# Best Subway Stations to Raise #StopAsianHate Awareness

I used Seaborn to plot a bar chart of the busiest subway stations in NYC from 10/10/2020 to 3/20/2021. The figure depicts the traffic for the top ten stations. The results suggest that 34 ST-PENN STATION is the busiest station for the past 6 months. 

Looking closer into the data, I took the top 5 busiest station and plotted the busiest time of the day (exlcuding late night from 10PM to 6AM). It seems like the best time to set up would be mornings at Flushing, Jackson Heights-Roosevelt Ave, 42 ST PORT AUTHORITY and mid-day at 34 ST-HERALD SQUARE AND GRAND CENTRAL-42 ST.
